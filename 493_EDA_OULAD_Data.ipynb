```python
# Fungsi untuk menghapus outlier berdasarkan IQR
def remove_outliers_iqr(df_studentVle, id_student):
    Q1 = df_studentVle[id_student].quantile(0.25)
    Q3 = df_studentVle[id_student].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data_filtered = df_studentVle[(df_studentVle[id_student] >= lower_bound) & (df_studentVle[id_student] <= upper_bound)]
    return data_filtered

# Menghapus outlier dari kolom 'id_student'
data_cleaned = remove_outliers_iqr(df_studentVle, 'id_student')

# Menyimpan hasil ke file CSV (opsional)
data_cleaned.to_csv('studentVle.csv', index=False)

# Menampilkan DataFrame yang sudah dihapus outlier
print(data_cleaned)
```

```python
#membuat dua kolom baru, 'unregistered' dan 'registered' untuk memisahkan siswa yang telah mendaftar dan tidak.
df_studentRegistration['unregistered'] = np.where(df_studentRegistration.date_unregistration=='?', 0, 1)
df_studentRegistration['registered'] = np.where(df_studentRegistration.date_unregistration=='?', 0, 1)
```

```python
#dataframe baru berisi informasi tentang jumlah siswa yang terdaftar dan tidak terdaftar berdasarkan ID siswa
student_interest = df_studentRegistration[['id_student', 'registered', 'unregistered']].groupby(['id_student']).sum()
student_interest = student_interest.reset_index()
student_interest.info()
```

```python
#visualisasi jumlah 'registered' dan 'unregistered'
student_interest[['registered', 'unregistered']].boxplot()
```

```python
# hitogram
student_interest.unregistered.hist()
```

```python
#mengubah nilai NaN menjadi 0
df_studentRegistration['register_days'] = (np.where(df_studentRegistration.date_registration=='?', 0,
                                          df_studentRegistration.date_registration)).astype(int)
df_studentRegistration['unregister_days'] = (np.where(df_studentRegistration.date_unregistration=='?', 0,
                                            df_studentRegistration.date_unregistration)).astype(int)
```

```python
# membuat dataFrame baru student_reg_days yang berisi nilai rata-rata dari 'register_days' dan 'unregister_days'
student_reg_days = df_studentRegistration[['id_student', 'register_days',
                                   'unregister_days']].groupby(['id_student']).mean()
student_reg_days = student_reg_days.reset_index()
student_reg_days.head()
```

```python
#histogram kolom 'unregister' setelah dipreprocessing
student_reg_days.unregister_days.hist()
```
